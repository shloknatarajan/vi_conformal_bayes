Overall Averages

```python
def results_summary(results):
    coverages = []
    for arr in results['coverage']:
        coverages.extend(arr)
    print(f"Coverage: {np.average(coverages)} +- {np.std(coverages)}")

    lengths = []
    for arr in results['length']:
        lengths.extend(arr)
    print(f"Lengths: {np.average(lengths)} +- {np.std(lengths)}")

    times = []
    for arr in results['time']:
        times.extend(arr)
    print(f"Times: {np.average(times)} +- {np.std(times)}")
    print()
```

```python
def results_summary_arrays(results):
    coverages = []
    for arr in results['coverage']:
        coverages.extend(arr)
    coverages_average = np.average(coverages)
    coverages_std = np.std(coverages)
    
    lengths = []
    for arr in results['length']:
        lengths.extend(arr)
    lengths_average = np.average(lengths)
    lengths_std = np.std(lengths)
    
    times = []
    for arr in results['time']:
        times.extend(arr)
    times_average = np.average(times)
    times_std = np.std(times)
    return [coverages_average, coverages_std, lengths_average, lengths_std, times_average, times_std]
```
